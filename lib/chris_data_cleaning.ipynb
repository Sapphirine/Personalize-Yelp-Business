# Business

```python
business_file = '/Users/chalpert/Documents/Columbia/Big_Data/Project/yelp_dataset_challenge_academic_dataset/yelp_academic_dataset_business.json'
```

```python
business = pd.read_json(business_file, orient='records', lines=True)
```

```python
def replace_chars(input_str):
    input_str = input_str.replace(' ', '_')
    input_str = input_str.replace(',', '_')
    input_str = input_str.replace(';', '_')
    input_str = input_str.replace('{', '_')
    input_str = input_str.replace('}', '_')
    input_str = input_str.replace('(', '_')
    input_str = input_str.replace(')', '_')
    input_str = input_str.replace('=', '_')
    return input_str

def replace_spaces(input_dict):
    new_dict = {}
    for old_key in input_dict:
        new_dict[replace_chars(old_key)] = input_dict[old_key]
    return new_dict
```

```python
business['attributes'] = business['attributes'].apply(replace_spaces)
```

```python
business.to_json('./business_cleaned.json', orient='records')
```

# Create business-category mapping

```python
def unpack_list(x, col):
    new_df = []
    
    records = x.to_dict(orient='records') 
    
    for i, website in enumerate(records):
        for c in website[col]:
            new_df.append(dict(website))
            new_df[-1]['%s_value' %col] = c
    return pd.DataFrame(new_df)
```

```python
business_category = unpack_list(business, 'categories')
```

```python
business_category[['business_id', 'categories_value']].to_json('./business_category.json', orient='records')
```
